## Basic simulation code

* `estimate_confint_t()` uses the t-distribution for estimating the confidence interval of the difference in means. See [`statsmodels` docs](http://www.statsmodels.org/stable/generated/statsmodels.stats.weightstats.CompareMeans.tconfint_diff.html) for details.
* `estimate_confit_arch()` is a thin wrapper around ARCH's `IndependentSamplesBootstrap.conf_int()` function, which makes it easy to test different bootstrapping-based algorithms for confidence interval estimation. See [`ARCH` docs](https://arch.readthedocs.io/en/latest/bootstrap/iid-bootstraps.html) for details.
* `generate_normal_samples()` is fairly self-explanatory: It generates samples from two independent normal distributions with mean zero and variance 1.
* `run_single_simulation()` is a general function for testing a confidence interval algorithm for a two-sample statistic. It calls the `generate_samples()` argument to generate the two samples with the given `simulation_i` seed, and then uses the `estimate_confint()` argument to estimate the confidence interval for the generated samples. It returns `'below'` if the `true_stat` argument is below the estimated confidence interval, `'above'` if it's above, and `'correct'` if it falls within the confidence interval.
* `check_confint_hit_rate()` runs `num_simulations` simulations by calling `run_single_simulation()` in a process pool. It returns a `dict` summarising the results, as shown by the examples below. You can adjust the number of CPUs used for simulations by changing the hardcoded `processes=8` value.

```python
def estimate_confint_t(sample_a, sample_b, _random_seed):
    return sm.stats.CompareMeans.from_data(sample_a, sample_b).tconfint_diff()


def estimate_confint_arch(sample_a, sample_b, random_seed,
                          reps=15000, method='percentile', studentize_reps=1000):
    bootstrap = IndependentSamplesBootstrap(sample_a, sample_b, random_state=np.random.RandomState(random_seed))
    return bootstrap.conf_int(lambda resample_a, resample_b: resample_b.mean() - resample_a.mean(),
                              reps=reps,
                              method=method,
                              studentize_reps=studentize_reps)


def generate_normal_samples(random_seed, size_a, size_b):
    rnd = np.random.RandomState(random_seed)
    return rnd.normal(size=size_a), rnd.normal(size=size_b)


def run_single_simulation(simulation_i, estimate_confint, generate_samples, true_stat):
    sample_a, sample_b = generate_samples(simulation_i)
    low, high = estimate_confint(sample_a, sample_b, simulation_i)
    if true_stat < low:
        return 'below'
    elif true_stat > high:
        return 'above'
    return 'correct'


def check_confint_hit_rate(estimate_confint, generate_samples, true_stat, num_simulations=1000):
    results = dict(correct=0, below=0, above=0)
    with multiprocessing.Pool(processes=8) as pool:
        for result in pool.map(functools.partial(run_single_simulation,
                                                 estimate_confint=estimate_confint,
                                                 generate_samples=generate_samples,
                                                 true_stat=true_stat),
                               range(num_simulations)):
            results[result] += 1
    return results
```

```python
check_confint_hit_rate(estimate_confint_t,
                       functools.partial(generate_normal_samples, size_a=10, size_b=20),
                       true_stat=0)
```

```python
check_confint_hit_rate(estimate_confint_arch,
                       functools.partial(generate_normal_samples, size_a=10, size_b=20),
                       true_stat=0)
```

## Conversion example

This example illustrates the problem with trying to infer the confidence interval for the difference between two samples from single-sample confidence intervals. Despite the overlap between the single-sample confidence intervals, the confidence interval for the difference in means doesn't contain zero.

```python
def plot_conversion_example(plot='clean', method='percentile'):
    sample_a = np.array([0] * 2150 + [1] * 350)
    sample_b = np.array([0] * 2090 + [1] * 410)
    diff_means = sample_b.mean() - sample_a.mean()
    ratio_means = sample_b.mean() / sample_a.mean()
    bootstrap = IndependentSamplesBootstrap(sample_a, sample_b, random_state=np.random.RandomState(0))
    def bootstrap_conf_int(func, add_point_estimate=False):
        ci = list(np.ravel(bootstrap.conf_int(func, reps=15000, method=method, studentize_reps=100)))
        if add_point_estimate:
            ci.insert(1, func(sample_a, sample_b))
            ci[0] = ci[1] - ci[0]
            ci[2] -= ci[1]
        return ci
    sample_a_ci = bootstrap_conf_int(lambda resample_a, _: resample_a.mean(), add_point_estimate=True)
    sample_b_ci = bootstrap_conf_int(lambda _, resample_b: resample_b.mean(), add_point_estimate=True)
    diff_means_ci = bootstrap_conf_int(lambda resample_a, resample_b: resample_b.mean() - resample_a.mean())
    ratio_means_ci = bootstrap_conf_int(lambda resample_a, resample_b: resample_b.mean() / resample_a.mean())
    df = pd.DataFrame(
        {
            'A': sample_a_ci,
            'B': sample_b_ci,
            'Difference\n(B - A)': [diff_means - diff_means_ci[0], diff_means, diff_means_ci[1] - diff_means],
            'Ratio\n(B / A)': [ratio_means - ratio_means_ci[0], ratio_means, ratio_means_ci[1] - ratio_means],
        },
        index=['bottom', 'mean', 'top']
    ) * 100
    if plot == 'clean':
        df_ab = df[['A', 'B']].copy()
        ax = df_ab.loc['mean'].plot.barh(xerr=df_ab.loc[['bottom', 'top']].values,
                                         rot=0,
                                         width=0.9,
                                         figsize=(8, 5))  
        df_ab.loc['bottom'] = -(df_ab.loc['bottom'] - df_ab.loc['mean'])
        df_ab.loc['top'] += df_ab.loc['mean']
    elif plot == 'diff':
        df_diff = df[['A', 'B', 'Difference\n(B - A)']].copy()
        ax = df_diff.loc['mean'].plot.barh(xerr=df_diff.loc[['bottom', 'top']].values,
                                           rot=0,
                                           width=0.9,
                                           figsize=(8, 5))
        df_diff.loc['bottom'] = -(df_diff.loc['bottom'] - df_diff.loc['mean'])
        df_diff.loc['top'] += df_diff.loc['mean']
    else:
        raise ValueError(f'Bad plot: {plot}')
    ax.invert_yaxis()
    ax.set_xlabel('Conversion rate [%]')
    ax.set_xlim(0, )
    ax.set_frame_on(False)
    ax.tick_params(axis='y', length=0)
    ax.grid(True, which='major', axis='x', linestyle=':')
    df.loc['bottom'] = -(df.loc['bottom'] - df.loc['mean'])
    df.loc['top'] += df.loc['mean']
    return df

plot_conversion_example('clean')
plt.figure()
plot_conversion_example('diff')
```

## Revenue example

This is a more elaborate example that applies in many freemium scenarios: We have several pricing tiers with different probabilities of uptake by users, including the special zero price (i.e., free). Due to different promotions and currency fluctuations, the revenue we get from each sale is somewhere around the advertised price. Therefore, we simulate the overall revenue distribution as:

* Draw sample sizes from `multinomial(total_size, weights)`, where `total_size` is the total number of sales (including free "sales") and `weights` is the vector of weights assigned to each price tier (i.e., their probabilities).
* For each non-zero price tier and its sample size `rnd_size`, draw `rnd_size` revenue instances from a Poisson distribution that has the advertised price as its mean.
* For the zero price tier and its sample size `rnd_size`, return `rnd_size` zeroes.

The results of this generative process for two different pricing schemes can be seen in the plots below.

```python
def generate_revenue_sample(rnd, mean_to_size):
    if isinstance(rnd, int):
        rnd = np.random.RandomState(rnd)
    sample = []
    total_size = sum(mean_to_size.values())
    weights = [size / total_size for size in mean_to_size.values()]
    true_mean = sum(mean * weight for mean, weight in zip(mean_to_size, weights))
    for mean, rnd_size in zip(mean_to_size.keys(), rnd.multinomial(total_size, weights)):
        if mean:
            sample.extend(rnd.poisson(mean, rnd_size))
        else:
            sample.extend([0] * rnd_size)
    return pd.Series(sample), true_mean


def plot_revenue_example(a_mean_to_size, b_mean_to_size):
    rnd = np.random.RandomState(0)
    sample_a, true_mean_a = generate_revenue_sample(rnd, a_mean_to_size)
    sample_b, true_mean_b = generate_revenue_sample(rnd, b_mean_to_size)
    assert len(sample_a) == len(sample_b)
    print(f'Sample A: {sample_a.mean()} (true: {true_mean_a})')
    print(f'Sample B: {sample_b.mean()} (true: {true_mean_b})')
    print(f'Diff (B - A): {sample_b.mean() - sample_a.mean()} (true: {true_mean_b - true_mean_a})')
    for ax in pd.DataFrame({'A': sample_a, 'B': sample_b}).plot.hist(subplots=True, bins=50):
        ax.set_xlim(0, 250)
        ax.set_ylim(0, 70)
        ax.set_ylabel('Count')
        ax.set_xlabel('Revenue [$]')

plot_revenue_example({0: 40, 25: 30, 50: 20, 100: 10},
                     {0: 60, 50: 25, 100: 10, 200: 5})
```

## Comparing different bootstrapping confidence intervals on revenue distributions

The code below simulates two different revenue distributions and checks how the accuracy of 95% confidence interval estimation varies by algorithm, sample size, and the number of bootstrapping resamples. As the figure shows, both the percentile and BCa algorithms perform poorly despite the BCa algorithm theoretically being more accurate. Even with a sample size of 10K, they both yield "95%" confidence intervals that contain the true difference in means less than 90% of the time, i.e., the intervals are too narrow.

The studentized algorithm gets much closer to the requested confidence level. However, this accuracy comes at the price of considerably longer runtime due to the need for nested bootstrapping. I've only used 10 resamples for the nested bootstraps (`studentize_reps` in the code) as I found in initial experiments that using 100 resamples doesn't substantially affect accuracy.

```python
def generate_specific_revenue_samples(random_seed, multiplier):
    a_mean_to_size = {0: 40 * multiplier,
                      25: 30 * multiplier,
                      50: 20 * multiplier,
                      100: 10 * multiplier}
    b_mean_to_size =  {0: 60 * multiplier,
                       50: 25 * multiplier,
                       100: 10 * multiplier,
                       200: 5 * multiplier}
    return (generate_revenue_sample(random_seed, a_mean_to_size)[0],
            generate_revenue_sample(random_seed, b_mean_to_size)[0])


def run_simulation_grid(methods=('percentile', 'bca', 'studentized_10'),
                        all_reps=(100, 1000, 10000),
                        num_simulations=1000,
                        sample_multiplier=1):
    df = pd.DataFrame(columns=methods, index=all_reps)
    for method in methods:
        for reps in all_reps:
            print(method, reps)
            if method.startswith('studentized'):
                actual_method, studentize_reps = method.split('_')
            else:
                actual_method = method
                studentize_reps = 0
            df.loc[reps, method] = check_confint_hit_rate(
                functools.partial(estimate_confint_arch,
                                  reps=reps,
                                  method=actual_method,
                                  studentize_reps=int(studentize_reps)),
                functools.partial(generate_specific_revenue_samples,
                                  multiplier=sample_multiplier),
                true_stat=5,
                num_simulations=num_simulations
            )['correct'] / num_simulations
    return df
```

## Time series example

Using toy data, we show the difference between IID and circular block bootstrapping. As the plot shows, blindly using IID bootstrapping for time series data isn't a great idea.

```python
def plot_time_series():
    T = 1000
    x = np.arange(0, T)
    y = np.sin(4 * np.pi * x / T) + np.cos(8 * np.pi * x / T)
    sample = pd.Series(index=x, data=y)
    rnd = np.random.RandomState(0)
    axes = pd.DataFrame({
        'Original': sample,
        'IID': next(IndependentSamplesBootstrap(sample, random_state=rnd).bootstrap(1))[0][0].values,
        'CircularBlock': next(CircularBlockBootstrap(100, sample, random_state=rnd).bootstrap(1))[0][0].values,
    }).plot(subplots=True)
    for ax in axes:
        ax.tick_params(length=0)
        ax.set_xlim(0, T)
        ax.set_xticks([])
        ax.set_ylim(-2.5, 1.5)
        ax.set_yticks([-2, -1, 0, 1])
        ax.legend(loc='lower left')

plot_time_series()
```
